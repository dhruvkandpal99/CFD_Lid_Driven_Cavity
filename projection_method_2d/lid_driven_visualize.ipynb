```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python

velocity_log = np.load('velocity_log.npy')  # Shape: (50, 129, 129)

fig, ax = plt.subplots(figsize=(6, 6))
im = ax.imshow(velocity_log[0], origin='lower', cmap='viridis', extent=[0, 1, 0, 1])
cbar = fig.colorbar(im, ax=ax)
cbar.set_label('Velocity Magnitude')

ax.set_title('Velocity Magnitude Evolution')
ax.set_xlabel('x')
ax.set_ylabel('y')

def update(frame):
    im.set_array(velocity_log[frame])
    ax.set_title(f'Iteration {frame}')
    return [im]

anim = animation.FuncAnimation(fig, update, frames=len(velocity_log), blit=True, interval=100)

# Save as MP4 using pillow
anim.save('velocity_evolution.mp4', writer='ffmpeg', fps=10)
```
